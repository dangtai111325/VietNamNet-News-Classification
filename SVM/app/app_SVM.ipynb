```python
# -- Kiểm tra môi trường ------------------------------------------------
import os, sys

NOTEBOOK_DIR  = os.getcwd()
PIPELINE_PATH = os.path.normpath(os.path.join(NOTEBOOK_DIR, '..', 'model', 'inference_pipeline.pkl'))

print('Kiểm tra thư viện:')
ok = True
missing_pkgs = []
for _pkg in ['streamlit','requests','bs4','pyvi','sklearn','numpy','pandas']:
    try:
        __import__(_pkg)
        print(f'  OK  {_pkg}')
    except ImportError:
        print(f'  MISS {_pkg}')
        missing_pkgs.append(_pkg)
        ok = False

print()
print('Kiểm tra model:')
if os.path.exists(PIPELINE_PATH):
    print(f'  OK  inference_pipeline.pkl  ({os.path.getsize(PIPELINE_PATH)/1e6:.1f} MB)')
else:
    print(f'  MISS {PIPELINE_PATH}')
    ok = False

if not ok:
    print()
    print('=' * 60)
    print('  KHÔNG THỂ TIẾP TỤC')
    print('=' * 60)
    if not os.path.exists(PIPELINE_PATH):
        print('  Model chưa được tạo.')
        print('  -> Chạy main_SVM.ipynb (Section 7) để export inference_pipeline.pkl trước.')
    if missing_pkgs:
        print('  Thiếu thư viện. Cài đặt:')
        print(f'     pip install {chr(32).join(missing_pkgs)}')
    print('=' * 60)
    raise SystemExit('Chua san sang -- xem huong dan o tren.')

print()
print('Sẵn sàng! Chạy cell tiếp theo để khởi động app.')
```

## ✍️ Cell 1 — Tạo / Cập nhật App

Kiểm tra file `app_SVM.py`. Chạy lại cell này bất cứ khi nào muốn xác nhận app đã sẵn sàng.

```python
import os

_app_file = os.path.join(os.getcwd(), "app_SVM.py")

# Cell này chỉ kiểm tra file tồn tại và các tính năng chính.
if os.path.exists(_app_file):
    _size = os.path.getsize(_app_file) / 1024
    print(f"✅ app_SVM.py tồn tại  ({_size:.1f} KB)")

    with open(_app_file, encoding="utf-8") as f:
        _src = f.read()
    if "get_top_keywords" in _src:
        print("✅ Top keywords: có trong app")
    else:
        print("⚠️  Top keywords: CHƯA có")
    if "batch" in _src.lower():
        print("✅ Nhiá»u URL: có trong app")
    else:
        print("⚠️  Nhiá»u URL: CHƯA có")
else:
    print(f"❌ Không tìm thấy: {_app_file}")
    print("   → Kiểm tra lại file SVM/app/app_SVM.py")

print("\n→ Chạy Cell 2 để khởi động app.")
```

## 🚀 Cell 2 — Khởi Động App

Mở **http://localhost:8501** trên trình duyệt.

> Cell này chạy cho đến khi bạn nhấn **■ Stop** (Interrupt Kernel).
> Hoặc mở terminal riêng trong thư mục này: `streamlit run app_SVM.py`

```python
import subprocess, sys, os, threading, webbrowser, time

_cmd = [sys.executable, "-m", "streamlit", "run", "app_SVM.py",
        "--server.port", "8501", "--server.headless", "true"]
print("Chạy:", " ".join(_cmd))
print("Mở trình duyệt: http://localhost:8501")
print("Nhấn ■ Interrupt Kernel để dừng.\n")

threading.Thread(
    target=lambda: (time.sleep(4), webbrowser.open("http://localhost:8501")),
    daemon=True
).start()

subprocess.run(_cmd, cwd=os.getcwd())
```
