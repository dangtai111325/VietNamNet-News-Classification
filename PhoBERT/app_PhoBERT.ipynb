|------|----------|
| **0. Kiểm tra** | Xác nhận thư viện và mô hình |
| **1. Tạo app** | Ghi code vào `app/main_app.py` |

> Sau mỗi lần train model xong, chạy lại **Cell 1 → Cell 2** là dùng được ngay.

```python
# -- Kiem tra moi truong ------------------------------------------------
import os, sys, json

MODEL_DIR    = os.path.normpath(os.path.join(os.getcwd(), 'model'))
CONFIG_PATH  = os.path.join(MODEL_DIR, 'label_config.json')
MODEL_FILE   = os.path.join(MODEL_DIR, 'model.safetensors')
THR_PATH     = os.path.join(MODEL_DIR, 'thresholds.json')

print('Kiem tra thu vien:')
ok = True
missing_pkgs = []
for _pkg in ['streamlit','requests','bs4','pyvi','torch','transformers',
             'numpy','pandas','scipy']:
    try:
        __import__(_pkg)
        print(f'  OK  {_pkg}')
    except ImportError:
        print(f'  MISS {_pkg}')
        missing_pkgs.append(_pkg)
        ok = False

print()
print('Kiem tra model:')
_required = [
    ('model.safetensors',     MODEL_FILE),
    ('label_config.json',     CONFIG_PATH),
    ('config.json',           os.path.join(MODEL_DIR, 'config.json')),
    ('tokenizer_config.json', os.path.join(MODEL_DIR, 'tokenizer_config.json')),
]
for _name, _path in _required:
    if os.path.exists(_path):
        print(f'  OK  {_name}  ({os.path.getsize(_path)/1e6:.1f} MB)')
    else:
        print(f'  MISS {_name}')
        ok = False

if os.path.exists(THR_PATH):
    print('  OK  thresholds.json -- Threshold calibration: BAT')
else:
    print('  WARN thresholds.json chua co -- app dung threshold mac dinh 0.5')

if not ok:
    print()
    print('=' * 60)
    print('  KHONG THE TIEP TUC')
    print('=' * 60)
    if not os.path.exists(MODEL_FILE):
        print('  Model chua duoc tao.')
        print('  -> Chay main_PhoBERT.ipynb (Section 7) de export model truoc.')
    if missing_pkgs:
        print('  Thieu thu vien. Cai dat:')
        print(f'     pip install {chr(32).join(missing_pkgs)}')
    print('=' * 60)
    raise SystemExit('Chua san sang -- xem huong dan o tren.')

print()
print('San sang! Chay cell tiep theo de khoi dong app.')
```

---
## ✍️ Cell 1 — Tạo / Cập nhật App

Ghi toàn bộ code Streamlit vào `app/main_app.py`. Chạy lại cell này bất cứ khi nào muốn cập nhật app.

```python
import os, shutil

_app_dir  = os.path.join(os.getcwd(), "app")
_app_file = os.path.join(_app_dir, "main_app.py")
os.makedirs(_app_dir, exist_ok=True)

# Không ghi đè bằng %%writefile — main_app.py đã được cập nhật đầy đủ
# (bao gồm threshold calibration). Cell này chỉ kiểm tra file tồn tại.
if os.path.exists(_app_file):
    _size = os.path.getsize(_app_file) / 1024
    print(f"✅ app/main_app.py tồn tại  ({_size:.1f} KB)")

    # Kiểm tra nhanh xem có threshold support không
    with open(_app_file, encoding="utf-8") as f:
        _src = f.read()
    if "THRESHOLD_PATH" in _src:
        print("✅ Threshold calibration: có trong app")
    else:
        print("⚠️  Threshold calibration: CHƯA có — copy lại từ app/main_app.py")
else:
    print(f"❌ Không tìm thấy: {_app_file}")
    print("   → Kiểm tra lại thư mục PhoBERT/app/main_app.py")

print("\n→ Chạy Cell 2 để khởi động app.")
```

## 🚀 Cell 2 — Khởi Động App

Mở **http://localhost:8501** trên trình duyệt.

> Cell này chạy cho đến khi bạn nhấn **■ Stop** (Interrupt Kernel).
> Hoặc mở terminal riêng: `streamlit run app/main_app.py`

```python
import subprocess, sys, os, threading, webbrowser, time

_cmd = [sys.executable, "-m", "streamlit", "run", "app/main_app.py",
        "--server.port", "8501", "--server.headless", "true"]
print("Chạy:", " ".join(_cmd))
print("Mở trình duyệt: http://localhost:8501")
print("Nhấn ■ Interrupt Kernel để dừng.\n")

threading.Thread(
    target=lambda: (time.sleep(4), webbrowser.open("http://localhost:8501")),
    daemon=True
).start()

subprocess.run(_cmd, cwd=os.getcwd())
```
